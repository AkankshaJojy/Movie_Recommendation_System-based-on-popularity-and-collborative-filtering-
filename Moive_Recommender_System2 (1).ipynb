```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
target_user = int(input(" Enter UserID: "))
num_recommendations = int(input("Enter number of recommendations (N): "))
similar_users_threshold = int(input("Enter threshold for similar users (k): "))
```

```python
target_user_ratings = data[data['userId'] == target_user]
target_user_ratings.head()
```

```python

user_ratings = data.pivot(index='userId', columns='movieId', values='rating').fillna(0)
user_similarity = cosine_similarity(user_ratings)
user_ratings.head()
```

```python
similar_users_indices = user_similarity[target_user - 1].argsort()[::-1][1:similar_users_threshold + 1]
similar_users_indices
```

```python
similar_users_ratings = user_ratings.iloc[similar_users_indices]
similar_users_movies = similar_users_ratings.columns[user_ratings.loc[target_user] == 0]
similar_users_ratings.head()
```

```python
recommendations = pd.DataFrame(similar_users_ratings[similar_users_movies].mean().sort_values(ascending=False))
recommendations.columns = ['average_rating']
recommendations.index.name = 'movieId'
recommendations.head()
```

```python

recommendations = pd.merge(recommendations, movies, on='movieId').head(num_recommendations)
recommendations.head()
```
